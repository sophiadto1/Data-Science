```python
x=np.linspace(0,10,50) #setting up an array
```

```python
m=3
b=7

y=m*x+b
```

```python
plt.plot(x,y,'.') #plotting trend line
```

```python
noise= 10*np.random.randn(50,)
```

```python
y=m*x+b+noise
```

```python
plt.plot(x,y,'.') #plotting points on graph
```

```python
m_guess=5 #plotting graph and inserting trendline
b_guess=15

y_guess=m_guess*x+b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.')
```

```python
loss=np.sum((y_guess-y)**2)
```

```python
for m_guess in np.linspace(-10,10,10): #plotting best guess of graphs
  for b_guess in np.linspace(-10,10,10):

    y_guess=m_guess*x+b_guess

    plt.plot(x,y_guess,'r.');
    plt.plot(x,y,'g.');
    plt.show()

    loss=np.sum((y_guess-y)**2)

    print(loss,m_guess,b_guess)
```

```python
best_loss=100000 #plotting best guess of graphs 
best_m=0 
best_b=0
for m_guess in np.linspace(-10,10,20):
  for b_guess in np.linspace(-10,10,20):

    y_guess=m_guess*x+b_guess

    loss=np.sum((y_guess-y)**2)

    if loss<best_loss:

      best_loss=loss
      best_m=m_guess
      best_b=b_guess

      plt.plot(x,y_guess,'r.');
      plt.plot(x,y,'g.');
      plt.show()

      print(loss,m_guess,b_guess)
```
